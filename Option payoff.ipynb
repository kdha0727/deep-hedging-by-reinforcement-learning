Usual imports

```python
import numpy as np
import matplotlib.pyplot as plt
```

My functions to define the payoffs at the maturity

```python
def C(K):
    S_T = np.linspace(0,200,num=201)
    return np.maximum(S_T - K, 0)

def P(K):
    S_T = np.linspace(0,200,num=201)
    return np.maximum(K - S_T, 0)

def S(K=None):
    return np.linspace(0,200,num=201)
```

Call with strike 100. Long position. Note that the price of the option is not included.

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(-P(100))
plt.ylim(-50,50)
plt.axvline(100, color='r', linestyle='--')
plt.annotate("strike price",(100 + 2,0 + 2))
plt.title("Covered Call (Stock + Naked Call)")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.show()
```

Short position of put with 90.

```python
plt.plot(C(100))
plt.ylim(-50,50)
plt.axvline(90, color='r', linestyle='--')
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.annotate("strike price",(90 + 2,-10 + -2))
plt.title("Married Put (Stock + Put)")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.show()
```

Bull call spread. Note that the payoff of this position is always positive, which means you have to pay at the beginning to have payoff at the maturity.

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(+C(90)-C(110)-10)
plt.axvline(90, color='r', linestyle='--')
plt.axvline(110, color='r', linestyle='--')
plt.annotate("low strike price",(90 + 2,-10 + -2))
plt.annotate("high strike price",(110 + 2,10 + 2))
plt.title("Bull Call Spread")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(-P(90)+P(110)-10)
plt.axvline(90, color='r', linestyle='--')
plt.axvline(110, color='r', linestyle='--')
plt.annotate("low strike price",(110 + 2,-10 + -5))
plt.annotate("high strike price",(90 + 2,10 + 2))
plt.title("Bear Put Spread")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(-P(110)+C(90)-S()+100)
plt.axvline(90, color='r', linestyle='--')
plt.axvline(110, color='r', linestyle='--')
plt.annotate("put strike price",(90 + 2,-10 + -2))
plt.annotate("call strike price",(110 + 2,10 + 2))
plt.title("Protective Collar")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(P(90)+C(90)-10)
plt.axvline(90, color='r', linestyle='--')
plt.annotate("strike price",(90 + 2,-10 + -2))
plt.title("Long Straddle")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(P(80)+C(120)-10)
plt.axvline(100, color='r', linestyle='--')
plt.annotate("strike price",(100 + 2,-10 + -5))
plt.title("Long Strangle")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(C(80)-C(100)-C(100)+C(120)-10)
plt.title("Long Call Butterfly Spread")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(C(70)-C(85)-C(115)+C(130)-5)
plt.title("Iron Condor")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```

```python
plt.axhline(0, color='k', linestyle='-', linewidth=1)
plt.plot(P(70)-P(100)-C(100)+C(130)+15)
plt.title("Iron Butterfly")
plt.xlabel("Stock Price ($)")
plt.ylabel("Profit of Loss ($)")
plt.ylim(-50,50)
plt.show()
```
